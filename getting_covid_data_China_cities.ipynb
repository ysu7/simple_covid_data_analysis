## 5. Replace the NaN values with the text 'not available'

```python
df.head()
```

# 6. Load only China's data

```python
df.query("`Country/Region`=='China'").head(3)
```

## 7. Update the index
- If the indexes dont start from 0, then reset it to start from 0.
- You may need a new dataframe.

```python
china = df[df["Country/Region"]=="China"]
china = china.reset_index(drop=True)
china.head()
```

## 8. Discard Country/region, Lat and Long columns
- Creat a new dataframe

```python
china = china.drop(['Country/Region', 'Lat', 'Long'], axis=1)
china.head()
```

# 9. Transpose the table. Provinces as columns and Cases as rows

```python
china = china.T
china.reset_index(drop=True, inplace=True) # drop the province/state from indexing
china.columns = china.iloc[0] # set the column headers as the names of province/state
china = china.drop(0) # drop the first row which are the names of province/state
china = china.reset_index(drop=True) # reset indexing after dropping the row
china.head()
```
